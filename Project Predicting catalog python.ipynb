# We define the Target variable and the Predcitor variables

```python
#Target Variable(Y)

Y= customer["Avg_Sale_Amount"]
display(Y)
```

```python
#Predictor variables(X)

X = customer.drop(["Customer_Segment_Credit Card Only","Avg_Sale_Amount","Name","Customer_ID","Address","City","State","ZIP","Store_Number","Responded_to_Last_Catalog","#_Years_as_Customer"], axis = 1)
```

```python
#Training the model
model = LinearRegression()
model.fit(X,Y)
```

```python
# Split dataset into training set and test set(70% and 30% respectively)

# Import train_test_split function

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3) # 70% training and 30% test
```

```python
#Fitting the model
model.fit(X_train, Y_train)
```

```python
#Bringing in our mailing list dataset for prediction
mailing_list = pd.read_excel(r'C:\Users\DELL\Downloads\p1-mailinglist.xlsx')
```

```python
#Coverting the Customer_segment to dummy variables
mailing_list = pd.get_dummies(mailing_list, columns=['Customer_Segment'])
```

```python
#Combining Score_Yes and Score_No into a column named Score

mailing_list["Score"] = mailing_list["Score_No"] + mailing_list["Score_Yes"]
```

```python
mail_test = mailing_list.drop(["Name","Customer_Segment_Credit Card Only","Customer_ID","Address","City","State","ZIP","Store_Number","Score_No","Score_Yes","#_Years_as_Customer","Score"], axis=1)
```

```python
#Predicting Profit Data

Predicted_Profit = model.predict(mail_test)
```

```python
#Adding the predicted Profit to the original mailing list Data

mailing_list['predicted_profit'] = Predicted_Profit
```

```python
mailing_list.head()
```

#We multiply the predicted profit by score_yes i.e. the probability that customer will respond to catalog

Score_Yes: The probability that the customer WILL respond to the catalog and make a purchase.

Hint: We want to calculate the expected revenue from these 250 people in order to get expected profit. This means we need to multiply the probability that a person will buy our catalog as well. For example, if a customer were to buy from us, we predict this customer will buy 450 dollars worth of products. At a 30 percent chance that this person will actually buy from us, we can expect revenue to be 450 x 30% = $135."



```python
mailing_list["expected_revenue"] = mailing_list["predicted_profit"] * mailing_list["Score_Yes"]
```

```python
mailing_list.head()
```

```python
#Taking the sum of all expected_Revenue for all the customers in the mailing list
Sum_Expected_Revenue = mailing_list["expected_revenue"].sum() 
```

```python
Sum_Expected_Revenue
```

# Calculating FInal Profit for all catalog for every customer in the mailing list
cost for printing all the catalog is 6.5 * 250 The average gross margin (price - cost) on all products sold through the catalog is 50%.

```python
Profit = (Sum_Expected_Revenue * 0.5) - (6.5 * 250)
```
